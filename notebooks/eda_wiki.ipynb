This notebook explores the words in the wiki file to get a sense of how to create a tokenizer

```python
file_path = "./data/text8"

with open(file_path, "r") as file:
    text = file.read()


# split the string into a list of words
list_of_words = text.split()

# lower case the words
list_of_words = [word.lower() for word in list_of_words]

# count the number of unique words using set list
unique_char = set(list_of_words)

# print the first 1000 words
print("first 1000 words:", text[:1500])
print("number of words in the text: ", len(list_of_words))
print("number of unique words in the text: ", unique_char)

```

    - by choosing the top N most frequent words.
